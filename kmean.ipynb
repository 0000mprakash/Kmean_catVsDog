```python
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from tqdm import tqdm
import os
import shutil
```

```python
def image_feature(direc):
    model = InceptionV3(weights='imagenet', include_top=False)
    features = [];
    img_name = [];
    for i in tqdm(direc):
        fname='D:\\Image Clustering\\Image Cluster bw img\\kmean_fea'+'/'+i
        img=image.load_img(fname,target_size=(224,224))
        x = img_to_array(img)
        x=np.expand_dims(x,axis=0)
        x=preprocess_input(x)
        feat=model.predict(x)
        feat=feat.flatten()
        features.append(feat)
        img_name.append(i)
    return features,img_name
```

```python
img_path=os.listdir('D:\\Image Clustering\\Image Cluster bw img\\kmean_fea')
```

```python
img_features,img_name=image_feature(img_path)
```

```python
image_cluster = pd.DataFrame(img_name,columns=['image'])
```

```python
image_cluster
```

```python
#Creating Clusters
k = 2
clusters = KMeans(k, random_state = 40)
clusters.fit(img_features)
```

```python
image_cluster["clusterid"] = clusters.labels_ # To mention which image belong to which cluster
```

```python
image_cluster # 0 denotes cat and 1 denotes dog
```

```python
os.mkdir('D:\\Image Clustering\\Image Cluster bw img\\kmean_cats')
os.mkdir('D:\\Image Clustering\\Image Cluster bw img\\kmean_dogs')
```

```python
# Images will be seperated according to cluster they belong
for i in range(len(image_cluster)):
    if image_cluster['clusterid'][i]==0:
        shutil.move(os.path.join('D:\\Image Clustering\\Image Cluster bw img\\kmean_fea', image_cluster['image'][i]), 'D:\\Image Clustering\\Image Cluster bw img\\kmean_dogs')
    else:
        shutil.move(os.path.join('D:\\Image Clustering\\Image Cluster bw img\\kmean_fea', image_cluster['image'][i]), 'D:\\Image Clustering\\Image Cluster bw img\\kmean_cats')
```
